## Importação das bibliotecas 

```python
import pandas as pd 
import sqlite3
import locale
```

## Lendo o Dataset de São Paulo 

```python
sao_paulo = pd.read_csv("saopaulo.csv", sep = ";")

sao_paulo
```

## Separação de Colunas 

```python
## Dim São Paulo Venddedor
Dim_Vendedor_Sao_Paulo = pd.DataFrame({
    
    "ID_Vendedor" : sao_paulo["ID_Vendedor"], 
    "Vendedor" : sao_paulo["Vendedor"]
})

Dim_Vendedor_Sao_Paulo
```

```python
## Dim São Paulo Produtos 
Dim_Produtos_Sao_Paulo = pd.DataFrame({
    
    "ID_Produto" : sao_paulo["ID_Produto"],
    "Fabricante" : sao_paulo["Fabricante"],
    "Categoria" : sao_paulo["Categoria"], 
    "Produto" : sao_paulo["Produto"],
    "Segmento" : sao_paulo["Segmento"]
    
})

Dim_Produtos_Sao_Paulo
```

```python
## Dim São Paulo Lojas 
Dim_Lojas_Sao_Paulo = pd.DataFrame({
    
    "Loja" : sao_paulo["Loja"], 
    "Cidade" : sao_paulo["Cidade"],
    "Estado" : sao_paulo["Estado"] 
})

Dim_Lojas_Sao_Paulo
```

```python
## Dim São Paulo Tempo 
Dim_Tempo_Sao_Paulo = pd.DataFrame({
    
    "Data_Venda" : sao_paulo["Data Venda"]
})

Dim_Tempo_Sao_Paulo
```

```python
## TB Fato 
TB_Fato_Sao_Paulo = pd.DataFrame({
    
    "Loja" : sao_paulo["Loja"], 
    "ID_Produto" : sao_paulo["ID_Produto"], 
    "ID_Vendedor" : sao_paulo["ID_Vendedor"], 
    "Data_Venda" : sao_paulo["Data Venda"],
    "Valor_Venda" : sao_paulo["ValorVenda"]
})

TB_Fato_Sao_Paulo
```

## Remoção de Dados Duplicados 

```python
## Dim_Vendedor_Sao_Paulo
Dim_Vendedor_Sao_Paulo = Dim_Vendedor_Sao_Paulo.drop_duplicates()
Dim_Vendedor_Sao_Paulo_count = Dim_Vendedor_Sao_Paulo.count()[0]

Dim_Vendedor_Sao_Paulo_count
```

```python
pd.DataFrame({"Total de Vendedores" : [Dim_Vendedor_Sao_Paulo_count]})
```

```python
Dim_Vendedor_Sao_Paulo
```

```python
## Dim São Paulo Lojas 
Dim_Lojas_Sao_Paulo = Dim_Lojas_Sao_Paulo.drop_duplicates() 
Dim_Lojas_Sao_Paulo_count = Dim_Lojas_Sao_Paulo.count()[0] 

Dim_Lojas_Sao_Paulo_count
```

```python
pd.DataFrame({"Total de Lojas" : [Dim_Lojas_Sao_Paulo_count]})
```

```python
Dim_Lojas_Sao_Paulo
```

## Conexão com o Banco de dados

```python
conn = sqlite3.connect("sp_regiao_vendas.db")
```

## Criação de Tabelas 

```python
## Criação da Tabela Dim_Vendedor_Sao_Paulo 
Dim_Vendedor_Sao_Paulo.to_sql(name='vendedores_sp', con=conn)
```

```python
## Criação da Tabela Dim_Lojas_Sao_Paulo
Dim_Lojas_Sao_Paulo.to_sql(name='lojas_sp', con=conn)
```

```python
## Criação da Tabela Dim_Produtos_Sao_Paulo
Dim_Produtos_Sao_Paulo.to_sql(name='produtos_sp', con=conn)
```

```python
## Criação da Tabela Dim_Tempo_Sao_Paulo
Dim_Tempo_Sao_Paulo.to_sql(name='tempo_sp', con=conn)
```

```python
## Criação da Tabela TB_Fato_São_Paulo
TB_Fato_Sao_Paulo.to_sql(name='fatos_sp', con=conn)
```

## 1 - Informe o total e média de vendas em São Paulo 

```python
consulta1 = '''
    
    SELECT ROUND(SUM(Valor_Venda)) AS TOTAL, ROUND(AVG(Valor_Venda)) AS Média FROM fatos_sp 

'''

consulta_total_media = pd.read_sql(consulta1, conn)

consulta_total_media
```

## 2 - Informe o total e média de Vendas por Categoria

```python
consulta2 = '''
    
    SELECT Categoria, SUM(Valor_Venda) AS Total, AVG(Valor_Venda) AS Média FROM fatos_sp f 
    INNER JOIN produtos_sp p ON f.ID_Produto = p.ID_Produto GROUP BY Categoria
'''

consulta_categoria = pd.read_sql(consulta2, conn)
```

```python
consulta_categoria
```

## 3 - Informe o total e média de vendas por Segmento e Estado 

```python
consulta3 = '''

    SELECT Segmento, Estado, SUM(Valor_Venda) AS Total, 
    AVG(Valor_Venda) AS Média FROM fatos_sp f 
    INNER JOIN lojas_sp l 
    ON f.Loja = l.Loja
    INNER JOIN produtos_sp p 
    ON f.ID_Produto = p.ID_Produto GROUP BY Estado, Segmento
    
'''

consulta_segmento_estado = pd.read_sql(consulta3, conn)
```

```python
consulta_segmento_estado
```

## 4 - Informe o total e média de vendas por Fabricante e Cidade

```python
consulta4 = '''
    
    SELECT Cidade, Fabricante, SUM(Valor_Venda) AS Total, 
    AVG(Valor_Venda) AS Média FROM fatos_sp f
    INNER JOIN lojas_sp l ON f.Loja = l.Loja 
    INNER JOIN produtos_sp p ON f.ID_Produto = p.ID_Produto 
    GROUP BY Fabricante
    
'''

consulta_fabricante_cidade = pd.read_sql(consulta4, conn)
```

## 5 - Informe o total, média e de vendas por Vendedor, Estado, Produto

```python
consulta5 = '''
    
    SELECT Vendedor, Estado, Produto, SUM(Valor_Venda) AS Total, AVG(Valor_Venda) AS Média
    FROM fatos_sp f 
    INNER JOIN vendedores_sp v ON f.ID_Vendedor = v.ID_Vendedor 
    INNER JOIN lojas_sp l ON f.Loja = l.Loja 
    INNER JOIN produtos_sp p ON f.ID_Produto = p.ID_Produto GROUP BY Vendedor
'''

consulta_vendedor_estado_produto = pd.read_sql(consulta5, conn)
```
