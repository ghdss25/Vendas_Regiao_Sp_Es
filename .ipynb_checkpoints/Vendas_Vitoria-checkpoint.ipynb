## Importação das bibliotecas 

```python
import pandas as pd 
import sqlite3
import locale
```

## Lendo o Dataset de São Paulo 

```python
vitoria = pd.read_csv("vitoria.csv", sep = ";")

vitoria
```

## Separação de Colunas 

```python
## Dim São Paulo Venddedor
Dim_Vendedor_Vitoria = pd.DataFrame({
    
    "ID_Vendedor" : vitoria["ID-Vendedor"], 
    "Vendedor" : vitoria["Vendedor"]
})

Dim_Vendedor_Vitoria
```

```python
## Dim São Paulo Produtos 
Dim_Produtos_Vitoria = pd.DataFrame({
    
    "ID_Produto" : vitoria["ID-Produto"],
    "Fabricante" : vitoria["Fabricante"],
    "Categoria" : vitoria["Categoria"], 
    "Produto" : vitoria["Produto"],
    "Segmento" : vitoria["Segmento"]
    
})

Dim_Produtos_Vitoria
```

```python
## Dim São Paulo Lojas 
Dim_Lojas_Vitoria = pd.DataFrame({
    
    "Loja" : vitoria["Loja"], 
    "Cidade" : vitoria["Cidade"],
    "Estado" : vitoria["Estado"] 
})

Dim_Lojas_Vitoria
```

```python
## Dim São Paulo Tempo 
Dim_Tempo_Vitoria = pd.DataFrame({
    
    "Data_Venda" : vitoria["Data Venda"]
})

Dim_Tempo_Vitoria
```

```python
## TB Fato 
TB_Fato_Vitoria = pd.DataFrame({
    
    "Loja" : vitoria["Loja"], 
    "ID_Produto" : vitoria["ID-Produto"], 
    "ID_Vendedor" : vitoria["ID-Vendedor"], 
    "Data_Venda" : vitoria["Data Venda"],
    "Valor_Venda" : vitoria["ValorVenda"]
})

TB_Fato_Vitoria
```

## Remoção de Dados Duplicados 

```python
## Dim_Vendedor_Sao_Paulo
Dim_Vendedor_Vitoria = Dim_Vendedor_Vitoria.drop_duplicates()
Dim_Vendedor_Vitoria_count = Dim_Vendedor_Vitoria.count()[0]

Dim_Vendedor_Vitoria_count
```

```python
pd.DataFrame({"Total de Vendedores" : [Dim_Vendedor_Vitoria_count]})
```

```python
Dim_Vendedor_Vitoria
```

```python
## Dim Vitoria Lojas 
Dim_Lojas_Vitoria = Dim_Lojas_Vitoria.drop_duplicates() 
Dim_Lojas_Vitoria_count = Dim_Lojas_Vitoria.count()[0] 

Dim_Lojas_Vitoria_count
```

```python
pd.DataFrame({"Total de Lojas" : [Dim_Lojas_Vitoria_count]})
```

```python
Dim_Lojas_Vitoria
```

## Conexão com o Banco de dados

```python
conn = sqlite3.connect("es_regiao_vendas.db")
```

## Criação de Tabelas 

```python
## Criação da Tabela TB_Fato_Vitoria
TB_Fato_Vitoria.to_sql(name='fatos_es', con=conn)
```

```python
## Criação da Tabela Dim_Lojas_Vitoria
Dim_Lojas_Vitoria.to_sql(name='lojas_es', con=conn)
```

```python
## Criação da Tabela Dim_Produtos_Vitoria
Dim_Produtos_Vitoria.to_sql(name='produtos_es', con=conn)
```

```python
## Criação da Tabela Dim_Tempo_Vitoria
Dim_Tempo_Vitoria.to_sql(name='tempo_es', con=conn)
```

```python
## Criação da Tabela Dim_Vendedor_Vitoria
Dim_Vendedor_Vitoria.to_sql(name='vendedores_es', con=conn)
```

```python
consulta1_es = '''
    
    SELECT ROUND(SUM(Valor_Venda)) AS TOTAL, ROUND(AVG(Valor_Venda)) AS Média FROM fatos_es 

'''

consulta_total_media_es = pd.read_sql(consulta1_es, conn)

consulta_total_media_es
```

## 2 - Informe o total e média de Vendas por Categoria

```python
consulta2_es = '''
    
    SELECT Categoria, SUM(Valor_Venda) AS Total, AVG(Valor_Venda) AS Média FROM fatos_es f 
    INNER JOIN produtos_es p ON f.ID_Produto = p.ID_Produto GROUP BY Categoria
'''

consulta_categoria_es = pd.read_sql(consulta2_es, conn)
```

## 3 - Informe o total e média de vendas por Segmento e Estado 

```python
consulta3_es = '''

    SELECT Segmento, Estado, SUM(Valor_Venda) AS Total, 
    AVG(Valor_Venda) AS Média FROM fatos_es f 
    INNER JOIN lojas_es l 
    ON f.Loja = l.Loja
    INNER JOIN produtos_es p 
    ON f.ID_Produto = p.ID_Produto GROUP BY Estado, Segmento
    
'''

consulta_segmento_estado_es = pd.read_sql(consulta3_es, conn)
```

## 5 - Informe o total, média e de vendas por Vendedor, Estado, Produto

```python
consulta5_es = '''
    
    SELECT Vendedor, Estado, Produto, SUM(Valor_Venda) AS Total, AVG(Valor_Venda) AS Média
    FROM fatos_es f 
    INNER JOIN vendedores_es v ON f.ID_Vendedor = v.ID_Vendedor 
    INNER JOIN lojas_es l ON f.Loja = l.Loja 
    INNER JOIN produtos_es p ON f.ID_Produto = p.ID_Produto GROUP BY Vendedor
'''

consulta_vendedor_estado_produto = pd.read_sql(consulta5_es, conn)
```
